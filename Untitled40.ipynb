```python
df['word count'] = df['text'].str.split().apply(len)
df
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

#feature_extractor = pipeline("feature-extraction", model="bert-base-uncased")
model = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model)

text = "I love you"
embedding = tokenizer(text)
embedding
#feature_extractor = AutoModelForSequenceClassification.from_pretrained(model)

```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(df,test_size=0.2)
test,val = train_test_split(test,test_size=0.5)

train.shape
test.shape
val.shape
```

```python
from datasets import Dataset, DatasetDict
dataset = DatasetDict(
    {'train':Dataset.from_pandas(train, preserve_index=False),
     'test':Dataset.from_pandas(test, preserve_index=False),
     'validation': Dataset.from_pandas(val, preserve_index=False)
     }

)

dataset
```

```python
def tokenize(data):
  temp = tokenizer(data['text'])
  return temp

tokenized_dataset = dataset.map(tokenize, batched=True)
```

```python
tokenized_dataset
```

```python
label2id = {x['label_name'] : x['label'] for x in dataset['train']}
id2label = {v : k for k,v in label2id.items()}

label2id
id2label
```

```python
from transformers import AutoModel
import torch
```

```python
model_name = "bert-base-uncased" # Use a new variable name for clarity
model = AutoModel.from_pretrained(model_name)
model
```

```python
from transformers import AutoModelForSequenceClassification, AutoConfig

num_labels = len(label2id)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
config = AutoConfig.from_pretrained(model_name, label2id=label2id, id2label=id2label)
model = AutoModelForSequenceClassification.from_pretrained(model_name, config=config).to(device)
```

```python
from transformers import TrainingArguments

batch_size = 64
training_dir = "bert_base_train_dir"

training_args = TrainingArguments( output_dir=training_dir,
                                  overwrite_output_dir = True,
                                  num_train_epochs = 2,
                                  learning_rate = 2e-5,
                                  per_device_train_batch_size = batch_size,
                                  per_device_eval_batch_size = batch_size,
                                  weight_decay = 0.01,
                                  disable_tqdm = False,
                                   report_to="none"
)
```

```python
from transformers import Trainer
trainer = Trainer(model = model,args = training_args, train_dataset = tokenized_dataset['train'], eval_dataset = tokenized_dataset['validation'], processing_class = tokenizer)
```

```python
trainer.train()
```

```python
trainer.save_model("bert-base-uncased-sentiment-model")
```
